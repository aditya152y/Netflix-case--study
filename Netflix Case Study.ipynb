## Analysis on Number of TV Shows and Movies

```python
plt.rcParams["figure.figsize"] = [5.00, 3.50]
plt.rcParams["figure.autolayout"] = True
ax = sns.countplot(x = 'type', data = netflix_copy)
plt.title("Comparison TV Shows and Movies in Netflix Library")
for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
n_Movie = netflix_copy[netflix_copy['type'] == 'Movie'].shape[0]
n_TV = netflix_copy[netflix_copy['type'] == 'TV Show'].shape[0]
data = [n_Movie, n_TV]
labels=['Movie','TV Show']
plt.title("Comparison TV Shows and Movies in Netflix Library")
plt.pie(data,
       labels=labels,
       autopct='%.0f%%') # To show the portions in %ages

plt.show()
```

# How has the number of movies released per year changed over past 20 - 30 Years

```python
movie1 = netflix_copy.loc[netflix_copy["type"] == "Movie" ,["release_year"]]
```

```python
movie1
```

```python
movie2 = movie1["release_year"].value_counts().sort_values()
```

```python
movie2
```

```python
movie3 = movie2.to_frame()
```

```python
movie3
```

```python
movie3.reset_index(inplace= True)
```

```python
movie3.columns = ["release_year" ,"movie_release_count"]
```

```python
movie3
```

```python
movie3_1 = movie3[movie3["release_year"] >= 1990]
```

```python
movie3_1
```

```python
ax1 = sns.lineplot(x= 'release_year', y= 'movie_release_count', data = movie3_1)
plt.title("Movie Release Scenario in last 30 Years")
plt.show()
```

# As we can see from above plot that there is a tremendous increase in movie release from year 1990 to 2015, i,e 16 movie to 398 movies per year
# The movie releases has decreased in 2020 by 250 from 2017-18 due to pandemic 

```python
movie3_1["release_percentage"] = movie3_1["movie_release_count"] / movie3_1["movie_release_count"].sum()
movie3_1["release_percentage"] = movie3_1["release_percentage"]*100
```

```python
movie3_1
```

```python
ax1 = sns.barplot(x= 'release_year', y= 'release_percentage', data = movie3_1)
plt.title("Movie Release Scenario in last 30 Years")
plt.xticks(rotation = 90)
plt.show()
```

```python
# 1.#The highest number of movies got released in 2017 and 2018 and the highest percentage increase in movie release is in 2016
# 2.The movie release in 2016 increased from 6% to approx 11%
```

```python
netflix_copy
```

```python
data1 = netflix_copy[netflix_copy["Year_added"] > 2010]
```

```python
movie5 = data1.loc[data1["type"]=="Movie",["Year_added"]].value_counts()
movie5 = movie5.to_frame()
movie5.reset_index(inplace = True)
movie5.columns = ["Year_added" ,"movie_added"]
```

```python
movie5.head(5)
```

```python
tvshow1 = data1.loc[data1["type"]=="TV Show",["Year_added"]].value_counts()
tvshow1 = tvshow1.to_frame()
tvshow1.reset_index(inplace = True)
tvshow1.columns = ["Year_added" ,"tvshow_added"]
```

```python
tvshow1
```

```python
movie_tv = movie5.merge(tvshow1, on="Year_added")
```

```python
movie_tv
```

```python
movie_tv["movie_added_percentage"] = movie_tv["movie_added"]/movie_tv["movie_added"].sum()
movie_tv["movie_added_percentage"] = movie_tv["movie_added_percentage"]*100
movie_tv["tvshow_added_percentage"] = movie_tv["tvshow_added"]/movie_tv["tvshow_added"].sum()
movie_tv["tvshow_added_percentage"] = movie_tv["tvshow_added_percentage"]*100
```

```python
movie_tv.sort_values(by="Year_added",inplace=True)
```

```python
movie_tv.plot(x="Year_added", y=["movie_added_percentage", "tvshow_added_percentage"], kind="bar")
plt.title("Comparison between Netflix focus on Movies and TV shows")
plt.show()
```
